```python
import os, shutil, random
DATASET = "/content/drive/MyDrive/fish_data"

base = DATASET

# source
img_src = os.path.join(base, "images")
lbl_src = os.path.join(base, "labels")

# target folders
for split in ["train", "val", "test"]:
    os.makedirs(os.path.join(base, "images", split), exist_ok=True)
    os.makedirs(os.path.join(base, "labels", split), exist_ok=True)

images = [f for f in os.listdir(img_src) if f.endswith((".jpg", ".png"))]
random.shuffle(images)

n = len(images)
train_split = int(0.8 * n)
val_split = int(0.9 * n)

for i, img in enumerate(images):
    label = img.replace(".jpg", ".txt").replace(".png", ".txt")

    if i < train_split:
        split = "train"
    elif i < val_split:
        split = "val"
    else:
        split = "test"

    shutil.move(os.path.join(img_src, img), os.path.join(base, "images", split, img))
    shutil.move(os.path.join(lbl_src, label), os.path.join(base, "labels", split, label))
```

```python
import os

DATASET = "/content/drive/MyDrive/fish_data"

yaml_content = f"""
path: {DATASET} # dataset root dir
train: images/train  # train images (relative to 'path')
val: images/val      # val images (relative to 'path')
test: images/test    # test images (relative to 'path')

# Classes
nc: 1  # number of classes (assuming a single class 'fish')
names: ['fish'] # class names
"""

# Write the YAML content to the fish.yaml file
fish_yaml_path = os.path.join(DATASET, "fish.yaml")
with open(fish_yaml_path, "w") as f:
    f.write(yaml_content)

print(f"fish.yaml created at {fish_yaml_path}")
```

```python
from ultralytics import YOLO

model = YOLO("/content/yolo12s.pt")

model.train(
    data="/content/drive/MyDrive/fish_data/fish.yaml",
    epochs=120,
    imgsz=768,
    batch=16,
    device="cpu",
    project="fish_detection",
    name="yolov12_fish",
    patience=75
)
```

```python
model = YOLO("fish_detection/yolov12_fish/weights/best.pt")

model.predict(
    source="/content/drive/MyDrive/fish_data/Samples",
    conf=0.35,
    save=True
)
```
